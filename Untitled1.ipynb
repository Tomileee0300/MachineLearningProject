```python


# ğŸ“š Import packages
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# ğŸ“‚ Upload your CSV file manually
from google.colab import files
uploaded = files.upload()

# ğŸ“„ Load data
df = pd.read_csv(next(iter(uploaded)))
df['date'] = pd.to_datetime(df['date'])
df['day_of_year'] = df['date'].dt.dayofyear
df['ghi'] = df['GHI']
df['dhi'] = df['DHI']

# â˜€ï¸ Estimate solar zenith angle
def solar_zenith_angle(day_of_year, lat=35):
    decl = 23.45 * np.sin(np.radians(360 * (284 + day_of_year) / 365))
    ha = 0
    zenith = np.arccos(
        np.sin(np.radians(lat)) * np.sin(np.radians(decl)) +
        np.cos(np.radians(lat)) * np.cos(np.radians(decl)) * np.cos(np.radians(ha))
    )
    return np.degrees(zenith)

df['zenith'] = df['day_of_year'].apply(solar_zenith_angle)

# ğŸ§® Compute DNI for each model
def compute_dni_models(row):
    dhi, ghi, zenith = row['dhi'], row['ghi'], row['zenith']
    cos_z = np.cos(np.radians(zenith))
    dni = {}
    dni['liu_jordan'] = (ghi - dhi) / max(cos_z, 1e-6)
    dni['koronakis'] = (ghi - dhi * (1 + np.cos(np.radians(zenith))) / 2) / max(cos_z, 1e-6)
    dni['badescu'] = (ghi - dhi * (1 + np.cos(np.radians(zenith))**2) / 2) / max(cos_z, 1e-6)
    dni['hay_davies'] = (ghi - dhi) / max(cos_z, 1e-6) + dhi * cos_z
    dni['reindl'] = (ghi - dhi * (1 + np.cos(np.radians(zenith)))) / max(cos_z, 1e-6)
    dni['hdkr'] = ((ghi - dhi) / max(cos_z, 1e-6)) + dhi * (1 + np.cos(np.radians(zenith)))
    return pd.Series(dni)

dni_df = df.apply(compute_dni_models, axis=1)
df = pd.concat([df, dni_df], axis=1)

# ğŸ§  Define neural network model
class SolarNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(3, 32),
            nn.ReLU(),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Linear(16, 1)
        )

    def forward(self, x):
        return self.model(x)

# ğŸ” Train model for each DNI output
features = df[['ghi', 'dhi', 'day_of_year']].values.astype(np.float32)
X_tensor = torch.tensor(features)

model_names = ['liu_jordan', 'koronakis', 'badescu', 'hay_davies', 'reindl', 'hdkr']
results = {}

for model_name in model_names:
    y = df[model_name].values.astype(np.float32)
    y_tensor = torch.tensor(y).view(-1, 1)

    dataset = TensorDataset(X_tensor, y_tensor)
    loader = DataLoader(dataset, batch_size=32, shuffle=True)

    model = SolarNet()
    loss_fn = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    # Train
    for epoch in range(100):
        for xb, yb in loader:
            pred = model(xb)
            loss = loss_fn(pred, yb)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

    # Predict
    model.eval()
    with torch.no_grad():
        y_pred_tensor = model(X_tensor)
        y_pred = y_pred_tensor.numpy().flatten()

    # Evaluate
    mse = mean_squared_error(y, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y, y_pred)

    results[model_name] = {
        'mse': mse,
        'rmse': rmse,
        'r2': r2,
        'preds': y_pred
    }

    # ğŸ“ˆ Plot
    plt.figure(figsize=(6, 4))
    plt.scatter(y, y_pred, alpha=0.4, s=10)
    plt.xlabel("Actual DNI")
    plt.ylabel("Predicted DNI")
    plt.title(f"{model_name.upper()} - Actual vs Predicted (NN)")
    plt.plot([min(y), max(y)], [min(y), max(y)], 'r--')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# ğŸ“‹ Show results
results_df = pd.DataFrame([
    {'Model': k, 'RMSE': v['rmse'], 'MSE': v['mse'], 'R2': v['r2']}
    for k, v in results.items()
])
print("\nğŸ“Œ PyTorch Neural Network Evaluation:")
display(results_df)
```

```python


# ğŸ“‚ Upload your CSV file
from google.colab import files
uploaded = files.upload()

# ğŸ“Š Import packages
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# ğŸ“„ Load and process data
df = pd.read_csv(next(iter(uploaded)))
df['date'] = pd.to_datetime(df['date'])
df['day_of_year'] = df['date'].dt.dayofyear
df['ghi'] = df['GHI']
df['dhi'] = df['DHI']

# â˜€ï¸ Zenith angle
def solar_zenith_angle(day_of_year, lat=35):
    decl = 23.45 * np.sin(np.radians(360 * (284 + day_of_year) / 365))
    ha = 0
    zenith = np.arccos(
        np.sin(np.radians(lat)) * np.sin(np.radians(decl)) +
        np.cos(np.radians(lat)) * np.cos(np.radians(decl)) * np.cos(np.radians(ha))
    )
    return np.degrees(zenith)

df['zenith'] = df['day_of_year'].apply(solar_zenith_angle)

# ğŸ”¬ DNI Models
def compute_dni_models(row):
    dhi, ghi, zenith = row['dhi'], row['ghi'], row['zenith']
    cos_z = np.cos(np.radians(zenith))
    dni = {}
    dni['liu_jordan'] = (ghi - dhi) / max(cos_z, 1e-6)
    dni['koronakis'] = (ghi - dhi * (1 + np.cos(np.radians(zenith))) / 2) / max(cos_z, 1e-6)
    dni['badescu'] = (ghi - dhi * (1 + np.cos(np.radians(zenith))**2) / 2) / max(cos_z, 1e-6)
    dni['hay_davies'] = (ghi - dhi) / max(cos_z, 1e-6) + dhi * cos_z
    dni['reindl'] = (ghi - dhi * (1 + np.cos(np.radians(zenith)))) / max(cos_z, 1e-6)
    dni['hdkr'] = ((ghi - dhi) / max(cos_z, 1e-6)) + dhi * (1 + np.cos(np.radians(zenith)))
    return pd.Series(dni)

dni_df = df.apply(compute_dni_models, axis=1)
df = pd.concat([df, dni_df], axis=1)

# ğŸ”„ Reverse GHI from DNI models
def reverse_ghi_models(row):
    dhi = row['DHI']
    zenith = row['zenith']
    cos_z = np.cos(np.radians(zenith))
    results = {}
    for model in ['liu_jordan', 'koronakis', 'badescu', 'hay_davies', 'reindl', 'hdkr']:
        dni = row[model]
        if model == 'liu_jordan':
            ghi = dhi + dni * cos_z
        elif model == 'koronakis':
            ghi = dhi * (1 + np.cos(np.radians(zenith))) / 2 + dni * cos_z
        elif model == 'badescu':
            ghi = dhi * (1 + np.cos(np.radians(zenith))**2) / 2 + dni * cos_z
        elif model == 'hay_davies':
            ghi = dni * cos_z + dhi * (1 - cos_z)
        else:
            ghi = dni * cos_z + dhi * (1 + np.cos(np.radians(zenith)))
        results[f'{model}_ghi'] = ghi
    return pd.Series(results)

ghi_df = df.apply(reverse_ghi_models, axis=1)
df = pd.concat([df, ghi_df], axis=1)

# ğŸ§  Multi-output Neural Network
class MultiOutputNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(3, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 6)
        )
    def forward(self, x):
        return self.net(x)

# ğŸ§® Prepare data
X = df[['DHI', 'hdkr', 'day_of_year']].values.astype(np.float32)
Y = df[[f'{m}_ghi' for m in ['liu_jordan', 'koronakis', 'badescu', 'hay_davies', 'reindl', 'hdkr']]].values.astype(np.float32)

X_tensor = torch.tensor(X)
Y_tensor = torch.tensor(Y)

dataset = TensorDataset(X_tensor, Y_tensor)
loader = DataLoader(dataset, batch_size=32, shuffle=True)

model = MultiOutputNet()
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# ğŸ” Train
for epoch in range(100):
    for xb, yb in loader:
        pred = model(xb)
        loss = loss_fn(pred, yb)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

# ğŸ“ˆ Evaluate
model.eval()
with torch.no_grad():
    preds = model(X_tensor).numpy()

# ğŸ“Š Metrics per output
model_names = ['liu_jordan', 'koronakis', 'badescu', 'hay_davies', 'reindl', 'hdkr']
results = []
for i, name in enumerate(model_names):
    actual = Y[:, i]
    predicted = preds[:, i]
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mse = mean_squared_error(actual, predicted)
    r2 = r2_score(actual, predicted)

    results.append({'Model': name, 'RMSE': rmse, 'MSE': mse, 'R2': r2})

    # Plot
    plt.figure(figsize=(6, 4))
    plt.scatter(actual, predicted, alpha=0.4, s=10)
    plt.xlabel("Actual GHI")
    plt.ylabel("Predicted GHI")
    plt.title(f"{name.upper()} - Actual vs Predicted")
    plt.plot([min(actual), max(actual)], [min(actual), max(actual)], 'r--')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# ğŸ“‹ Summary
import pandas as pd
results_df = pd.DataFrame(results)
print("\nğŸ“Š Multi-Output Neural Network Evaluation:")
display(results_df)
```
